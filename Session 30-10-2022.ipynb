Operators

Arithematic Operator

```python
print( 5 + 5 )
print( 5 - 5 )
print( 5 * 5 )
print( 5 / 3 )
print( 5 ** 2 ) # this means 5 raised to the power 2
print( 5 // 3 ) #this will give the quotient when 5 is divided by 3 (floor division operator)
print( 5 % 3 ) #this will give the remainder when 5 is divided by 3 (modulous/modulus/modulo operator)
```

Comparison Operator

```python
print( 6 > 6 )
print( 6 >= 6 )
print( 6 < 5 )
print( 6 <= 5 )
print( 6 == 5 ) #this is used to compare the two number for their equlity 
print( 6 != 5 )
var1 = 6 #this means you are assigning 6 to var1
var1 == 6 #wheras this means you are comparing whether var1 is having the valule equals to 6
```

Logical Operator

```python
print( 6 > 5 and 7 < 4) #and operator will result in true only when both the statements are true
print( 6 > 5 or 7 < 4) #or operator will result in true when either of the statements are true
```

Conditional Statements

if statement

```python
#WAP to find whether a number is even

num = int(input("Enter a number: "))

print(f"You have entered -> {num}")
if( num % 2 == 0 ):
    #you are in the block of if statement
    print(f"{num} is an even number")

#now you are out of the if block
print("Program executed normally")
```

if-else statement

```python
#WAP to find whether a number is even or odd

num = int(input("Enter a number: "))

print(f"You have entered -> {num}")
if( num % 2 == 0 ):
    #you are in the block of if statement
    print(f"{num} is an even number")

else:
    #you are in the block of else statement
    print(f"{num} is an odd number")

#now you are out of the if block
print("Program executed normally")
```

2. WAP to find the simple interest on a principal amount for a period time at certain rate of interest enetered by the user and if the interest amount is greater than 5000 then deduct a tds of 10% from it and show the final amount to the user and tds dedcuted if any.

1. Simple Interest = Principal * rate of Interest * Time / 100
2. Amount = Amount + Simple Interest after tds deduction

```python
principal = int(input("Enter the principal amount: "))
roi = float(input("Enter rate of interest: "))
time = int(input("Enter time time period: "))

print(f"Principal amount = {principal}, Rate of Interest = {roi}, Time = {time}")

si = principal * roi * time / 100
tds = 0

if( si > 5000 ):
    tds = si * 0.1
    si = si - tds #this is nothing but si -= tds
    #si *= tds -> si = si * tds

final_amount = principal + si
print(f"Final amount is {final_amount} and tds deducted = {tds} and Simple Interest = {si}")
```

if-elif-else statement

```python
num = int(input("Enter a number: "))

if (num > 50):
    print(f"{num} is greater than 50")

elif ( num == 50 ): #here elif means else if
    print(f"{num} is equals to 50")

else:
    print(f"{num} is less than 50")
```

1. WAP to calculate the tax on income accroding to following logic - 
2. 0 - 250000 -> Nil
3. 250001 - 500000 -> 5%
4. 500001 - 750000 -> 10%
5. 750001 - 1000000 -> 15%
6. 1000001 - 1250000 -> 20%
7. 1250001 - 1500000 -> 25%

Note - The above tax calculation is not as per income tax rules this is just for easiness to apply the logic for elif statement

```python
income = int(input("Enter your income: "))
if( income >= 0 and income <=250000 ):
    print(f"Tax on an income of Rs. {income} is Rs. 0")
elif( income >= 250001 and income <=500000 ):
    print(f"Tax on an income of Rs. {income} is Rs. {income * 0.05}")
elif( income >= 500001 and income <= 750000 ):
    print(f"Tax on an income of Rs. {income} is Rs. {income * 0.1}")
elif( income >= 750001 and income <= 1000000 ):
    print(f"Tax on an income of Rs. {income} is Rs. {income * 0.15}")
elif( income >= 1000001 and income <= 1250000 ):
    print(f"Tax on an income of Rs. {income} is Rs. {income * 0.2}")
elif( income >= 1250001 and income <= 1500000 ):
    print(f"Tax on an income of Rs. {income} is Rs. {income * 0.25}")
else:
    print(f"Tax on an income of Rs. {income} is Rs. {income * 0.3}")
```
